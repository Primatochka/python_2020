```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('horse_data.csv',header=None)

# какие типы признаков в нашем датафрейме?
df.head()
```

```python
df.info()
```

```python
df_1 = pd.DataFrame()
df_1['is_surgery'] = df.iloc[:, 0]
df_1['pulse'] = df.iloc[:, 4]
df_1['respiratory_rate'] = df.iloc[:, 5]
df_1['pain'] = df.iloc[:, 10]
df_1['abdominal_distension'] = df.iloc[:, 12]
df_1['protein'] = df.iloc[:, 19]
df_1['outcome'] = df.iloc[:, 22]
df_1['cp_data'] = df.iloc[:, 27]
```

```python
df_1.info()
```

```python
df_1.head()
```

```python
#расчитываем количество пропусков
df_1.isin(['?']).sum(axis=0)
```

```python
# в данных столбцах по одному пропуску. можно удалить строки с этими пропусками из датафрейма без потери качества выборки
df_1 = df_1[df_1.is_surgery != '?']
df_1 = df_1[df_1.outcome != '?']
```

```python
#df_1.loc[(df_1['is_surgery'] == '?'), 'is_surgery'] = 2
for i in list(df_1.columns):
    df_1.loc[(df_1[i] == '?'), i] = np.nan
    df_1[i] = pd.to_numeric(df_1[i], downcast="float")
```

```python
df_1.info()
```

```python
for i in list(df_1.columns):
    print(i)
    print(f'mean {df_1[i].mean()} - median {df_1[i].median()} - mode {df_1[i].mode()[0]} - {df_1[i].std()} - {df_1[i].var()}')
    print('------------********-----------')
```

Выводы:
для категориальных метрик медиана и мода обычно совпадают. Соответсвенно можно любое из этих значений использовать для замены пропусков.

```python
#для остальных столбцов формируем подмены по следующей логике:
#количественные значения - по медиане , качественные по моде
df_1.pulse.fillna(df_1.pulse.median(), inplace=True)
df_1.respiratory_rate.fillna(df_1.respiratory_rate.median(), inplace=True)
df_1.pulse.fillna(df_1.pulse.median(), inplace=True)
df_1.pain.fillna(3, inplace=True)
df_1.abdominal_distension.fillna(1, inplace=True)
df_1.protein.fillna(df_1.protein.mode()[0], inplace=True)
df_1.outcome.fillna(1, inplace=True)
```

```python
#датафрейм без пропусков
df_1.describe()
```

```python
df_1.isna().sum(axis=0)
```

Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
q1 = df_1.pulse.quantile(0.25)
q3 = df_1.pulse.quantile(0.75)
iqr = q3 - q1
print(q1, q3, iqr)
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
print(lower_bound)
print(upper_bound)
remove_outliers = df_1[df_1.pulse.between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse', ascending = False)
remove_outliers
```

```python
print(remove_outliers.pulse.mean())
print(remove_outliers.pulse.median())
```

```python
print(df_1.pulse.mean() - remove_outliers.pulse.mean())
```

```python
print(df_1.pulse.median() - remove_outliers.pulse.median())
```

```python
df_1[df_1.pulse >= 147.25].sort_values('outcome', ascending = False)
```
