```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_happy = pd.read_csv('archive/2019.csv')
```

```python
df_happy.head(10)
```

```python
df_happy.info()
```

```python
df_happy_1 = df_happy.sort_values(by='Perceptions of corruption', ascending=False).head(10)
df_happy_2 = df_happy.sort_values(by='Perceptions of corruption', ascending=True).head(10)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Индекс восприятия коррупции 2019')
fig.set_size_inches(18.5, 5.5)
ax1.plot(df_happy_1['Country or region'], df_happy_1['Perceptions of corruption'])
ax1.tick_params(labelrotation=45)
ax2.plot(df_happy_2['Country or region'], df_happy_2['Perceptions of corruption'])
ax2.tick_params(labelrotation=45)
ax1.set_title('топ 10 стран с минимальным уровнем')
ax2.set_title('топ 10 стран с максимальным уровнем')
```

Вывод по визуализации:
топ 10 стран с минимальным уровнем корупции на 50% состоит из северо-европейских стран
топ 10 стран с максимальным уровнем корупции на 50% состоит из восточно-европейских стран

```python
df_happy_3 = df_happy.sort_values(by='Overall rank', ascending=True).head(10)
df_happy_3 = df_happy_3.drop(columns=['Overall rank', 'Score'])
df_happy_4 = df_happy_3.groupby('Country or region').sum()
df_happy_4.plot(kind='bar', stacked=True)
```

```python
df_happy_5 = df_happy.sort_values(by='Overall rank', ascending=False).head(10)
df_happy_5 = df_happy_5.drop(columns=['Overall rank', 'Score'])
df_happy_6 = df_happy_5.groupby('Country or region').sum()
df_happy_6.plot(kind='bar', stacked=True)
```

Вывод по визуализации:
Большой вес для "уровня счастья" граждан страны имеют параметры: GDP per capita и Social support

```python
df_happy.plot(kind='scatter', 
        x='Social support', 
        y='Healthy life expectancy', 
        title='Зависимость между социальной поддержкой и продолжительностью жизни ')
plt.show()
```

Вывод по визуализации:
Большая социальная поддержка увеличивает продолжительность жизни
